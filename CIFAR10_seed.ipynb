```python
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.losses import sparse_categorical_crossentropy
from keras.optimizers import Adam
import matplotlib.pyplot as plt
import os
import time
from csv import writer
from csv import reader
import tensorflow as tf
import pandas as pd
from numpy.random import seed
import keras
import keras.backend as K
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
seeds = list(range(0,10))
multiply= [2]
class TimeHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.times = []

    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_time_start = time.time()

    def on_epoch_end(self, epoch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)
```

```python
# Model configuration
batch_size = 50
img_width, img_height, img_num_channels = 32, 32, 3
loss_function = sparse_categorical_crossentropy
no_classes = 10
no_epochs = 50
validation_split = 0.2
verbosity = 1
```

```python
# Load CIFAR-10 data
(input_train, target_train), (input_test, target_test) = cifar10.load_data()
```

```python
data_slice1 = 25000
data_slice2 = 15000
input_train = input_train[:data_slice1,:]
target_train = target_train[:data_slice1,:]
input_test = input_test[:data_slice2,:]
target_test = target_test[:data_slice2,:]
```

```python
# Determine shape of the data
input_shape = (img_width, img_height, img_num_channels)
```

```python
# Parse numbers as floats
input_train = input_train.astype('float32')
input_test = input_test.astype('float32')
```

```python
# Normalize data
input_train = input_train / 255
input_test = input_test / 255
```

```python
for mltp in multiply:
    for actual_seed in seeds:
        seed(actual_seed)
        tf.random.set_seed(actual_seed)
        random_state = actual_seed

        model = Sequential()
        model.add(Conv2D(32*mltp, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Conv2D(64*mltp, kernel_size=(3, 3), activation='relu'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Conv2D(128*mltp, kernel_size=(3, 3), activation='relu'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Flatten())
        model.add(Dense(256*mltp, activation='relu'))
        model.add(Dense(128*mltp, activation='relu'))
        model.add(Dense(no_classes, activation='softmax'))
        # Compile the model

        filename =".\CIFAR10_seed_result/"+"CIFAR10_" + str(mltp) + "_" +str(actual_seed)+ ".csv"
        history_logger= keras.callbacks.CSVLogger(filename, separator=",", append=True)
        time_callback = TimeHistory()
        model.compile(loss=loss_function,
                    optimizer="adam")

        # Fit data to model
        history = model.fit(input_train, target_train,
                    batch_size=batch_size,
                    epochs=no_epochs,
                    verbose=verbosity,
                    validation_split=validation_split,
                    callbacks=[history_logger,time_callback])
        
        times = time_callback.times
        csv_input = pd.read_csv(filename)
        csv_input['time'] = times
        csv_input.to_csv(filename, index=False)
        
        tf.keras.backend.clear_session()
        K.clear_session()
        del model
        tf.compat.v1.reset_default_graph()
```
